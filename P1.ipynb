## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    global FLAG_FOR_VIDEO # This flag is used to identify whether the input to the pipeline is image of a video (frames on images)
    global negslope_prev_frame
    global xtopneg_prev_frame
    global xbotneg_prev_frame
    global ytopneg_prev_frame
    global xtopnewneg_prev_frame

    global posslope_prev_frame
    global xtoppos_prev_frame
    global xbotpos_prev_frame
    global ytoppos_prev_frame
    global xtopnewpos_prev_frame

    yposmid = [] # list to store the mid point of the X and Y coordinates of the lines with positive slope;
    xposmid = []
    ynegmid = []
    xnegmid = []
    positiveslopes = []
    negativeslopes = []
    prev_posslope = 0

    for line in lines:
        for x1,y1,x2,y2 in line:
            slo = ((y2-y1)/(x2-x1))
            if (0.5 < slo < 0.9):
                if (x1 > 400 and x2 > 400):
                    positiveslopes.append(slo)
                    yposmid.append((y1+y2)/2)
                    xposmid.append((x1+x2)/2)

            elif( slo < -0.5 ):
                if (x1 < 400 and x1 > 100):
                    negativeslopes.append(slo)
                    ynegmid.append((y1+y2)/2)
                    xnegmid.append((x1+x2)/2)

# Right Lane                     
                    
# Find the index of top most and the bottom most Y coordinate and use it to find the corrresponding X coordinate 
    yposmax = np.amax(yposmid)
    indyposmax = yposmid.index(yposmax)
    xposmax = xposmid[indyposmax]

    yposmin = np.amin(yposmid)
    indyposmin = yposmid.index(yposmin)
    xposmin = xposmid[indyposmin]

# Use median of all the slopes to calculate the intercept
    posslope = np.median(positiveslopes)
    xtoppos = xposmin
    ytoppos = yposmin
    interceptpos = ytoppos - posslope*xtoppos
    
# Calculate the extrapolated X coordinates 

    xbotpos = (Ybot - interceptpos)/posslope
    xtopnewpos = (Ytop - interceptpos)/posslope

# LEFT Lane 

    ynegmax = np.amax(ynegmid)
    indynegmax = ynegmid.index(ynegmax)
    xnegmax = xnegmid[indynegmax]

    ynegmin = np.amin(ynegmid)
    indynegmin = ynegmid.index(ynegmin)
    xnegmin = xnegmid[indynegmin]

    negslope = np.median(negativeslopes)
    ytopneg = ynegmin
    xtopneg = xnegmin
    interceptneg = ytopneg - negslope*xtopneg
    xbotneg = (Ybot - interceptneg)/negslope
    xtopnewneg = (Ytop - interceptneg)/negslope

    if (FLAG_FOR_VIDEO == 0):   # if the input is just one image
         final_image = cv2.line(img,(int(xtopnewpos),int(Ytop)),(int(xbotpos),Ybot),(255,0,0),10)
         final_image = cv2.line(img,(int(xtopnewneg),int(Ytop)),(int(xbotneg),Ybot),(255,0,0),10)

    if (FLAG_FOR_VIDEO !=0): # if the input is stream of images
        # Calculating the difference (delta) between the previous frame and current
        # frame and multiplying it with a factor (to finetune the delta). Adding
        # this delta to the current frame
         posslope = (posslope_prev_frame - posslope)*0.9 + posslope
         xtoppos = (xtoppos_prev_frame - xtoppos)*0.9 + xtoppos
         ytoppos = (ytoppos_prev_frame - ytoppos)*0.9 + ytoppos
         interceptpos = ytoppos - posslope*xtoppos
         xbotpos = ((Ybot - interceptpos)/posslope - xbotpos)*0.9 + xbotpos
         xtopnewpos = ((Ytop - interceptpos)/posslope -xtopnewpos)*0.9 + xtopnewpos

         negslope = (negslope_prev_frame - negslope)*0.9 + negslope
         xtopneg = (xtopneg_prev_frame - xtopneg)*0.9 + xtopneg
         ytopneg = (ytopneg_prev_frame - ytopneg)*0.9 + ytopneg
         interceptneg = ytopneg - negslope*xtopneg
         xbotneg = ((Ybot - interceptneg)/negslope - xbotneg) + xbotneg
         xtopnewneg = ((Ytop - interceptneg)/negslope -xtopnewneg)*0.9 + xtopnewneg

         final_image = cv2.line(img,(int(xtopnewpos),int(Ytop)),(int(xbotpos),Ybot),(255,0,0),10)
         final_image = cv2.line(img,(int(xtopnewneg),int(Ytop)),(int(xbotneg),Ybot),(255,0,0),10)

# for the next iteration assigning the value of variables (*_prev_frame) that are used to
# calculate the difference between current frame and previous frame
    posslope_prev_frame = posslope
    xtoppos_prev_frame = xposmin
    ytoppos_prev_frame = yposmin
    xbotpos_prev_frame = xbotpos
    xtopnewpos_prev_frame = xtopnewpos

    negslope_prev_frame = negslope
    xtopneg_prev_frame  = xnegmin
    xbotneg_prev_frame = xbotneg
    ytopneg_prev_frame = ynegmin
    xtopnewneg_prev_frame = xtopnewneg
    
    FLAG_FOR_VIDEO = 1

    return final_image



def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
  #  draw_lines(line_img, lines)
    return lines


def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `intiail_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
Ytop = 320 # Top Y coordinate of the bounding box
Ybot = 550 # Bottom Y coordinate of the bounding box
Xtop_left = 475 # Top left coordinate of the bounding box
Xtop_right = 500 # Top right coordinate of the bounding box

# Flag to differentiate the image and the video (stream of images)
# For the first frame of the video stream FLAG_FOR_VIDEO = 0, else
# FLAG_FOR_VIDEO = 1
global FLAG_FOR_VIDEO

# GLobal variables to store the slope and corrdinates --> used to calculate the difference between the current and previous
# frame
global negslope_prev_frame
global xtopneg_prev_frame
global xbotneg_prev_frame
global ytopneg_prev_frame

global posslope_prev_frame
global xtoppos_prev_frame
global xbotpos_prev_frame
global ytoppos_prev_frame
global xtopnewpos_prev_frame

negslope_prev_frame = 0
xtopneg_prev_frame  = 0
xbotneg_prev_frame  = 0
ytopneg_prev_frame  = 0
xtopnewneg_prev_frame  = 0

posslope_prev_frame = 0
xtoppos_prev_frame  = 0
xbotpos_prev_frame  = 0
ytoppos_prev_frame  = 0
xtopnewpos_prev_frame  = 0

FLAG_FOR_VIDEO = 0
def pipeline(image):
    imshape = image.shape
    kernel_size = 5
    low_threshold = 30
    high_threshold = 100
    ignore_mask_color = 255

    vertices = np.array([[(0,imshape[0]),(Xtop_left, Ytop), (Xtop_right, Ytop), (imshape[1],imshape[0])]], dtype=np.int32)
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 8 # minimum number of votes (intersections in Hough grid cell)
    min_line_len = 20 #minimum number of pixels making up a line
    max_line_gap = 3 # maximum gap in pixels between connectable line segments

    line_image = np.copy(image)*0 # creating a blank to draw lines on
    gray = grayscale(image)
    blur_gray = gaussian_blur(gray,kernel_size)
    edges = canny(blur_gray,low_threshold,high_threshold)
    mask = np.zeros_like(edges)
    masked_edges = region_of_interest(edges,vertices)
    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)
    hough_image = np.zeros((*masked_edges.shape, 3), dtype=np.uint8)
    testneg = draw_lines(hough_image, lines, color=[255, 0, 0], thickness=2)
    lines_edges = weighted_img(testneg, image, 0.5, 0.9, 1.0)
    return lines_edges

````
